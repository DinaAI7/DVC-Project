```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("C:/Users/user/Downloads/netflix_titles.csv")

# Display the shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Get general information about the dataset
df.info()

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
missing_values = df.isna().sum()
print("Missing values in each column:\n", missing_values)

# Drop columns with all missing values
df = df.dropna(axis=1, how='all')

# Display the updated info of the dataframe
df.info()

# Handling missing values in specific columns (e.g., 'director' and 'cast')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')

# For columns like 'country', 'date_added', and 'rating', we can fill missing values with the mode
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])

# Verify that there are no missing values left
print("Missing values after cleaning:\n", df.isna().sum())
```

```python
import pandas as pd
from dateutil import parser

# Function to parse dates
def parse_date(date_str):
    try:
        return parser.parse(date_str)
    except:
        return pd.NaT

# Apply the function to 'date_added' column
df['date_added'] = df['date_added'].apply(parse_date)

# Check the result
df.info()
```

```python
# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'])

# Extract the year from 'date_added'
df['year_added'] = df['date_added'].dt.year

# Convert 'release_year' to numeric
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicate rows if any
df = df.drop_duplicates()

# Verify that there are no duplicate rows left
duplicates_after = df.duplicated().sum()
print(f"Number of duplicate rows after dropping: {duplicates_after}")
```

```python
# Display the cleaned dataset
df.head()

# Save the cleaned dataset to a new CSV file
df.to_csv('C:/Users/user/Downloads/cleaned_netflix_titles.csv', index=False)
```
